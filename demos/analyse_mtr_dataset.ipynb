## 3. Compute the number of points in the bounding boxes

1. Number of points before denoising raw

```python

save_viz_path = "/home/tan/tjtanaa/det3d/demos/visualization/mtr_dataset_inspection"
# Initialize and setup output directory.
Converter = PointvizConverter(save_viz_path)

from det3d.mtr_dataset.utils import mtr_utils

total_number_of_points = []
# total_number_of_objects = sum(total_number_of_objects_per_scene)
total_number_of_objects_per_scene = []
total_gt_bbox_parameter_list = []

for sample_id in trange(len(train_dataset)):
    point_cloud = train_dataset.get_lidar(sample_id)

    pts_features = point_cloud[:, 3:]

    obj_list = train_dataset.get_label(sample_id) # are labels

    gt_boxes3d = np.zeros((obj_list.__len__(), 7), dtype=np.float32)
    gt_bbox_params_list = []
    for k, obj in enumerate(obj_list):
        gt_boxes3d[k, 0:3], gt_boxes3d[k, 3], gt_boxes3d[k, 4], gt_boxes3d[k, 5], gt_boxes3d[k, 6] \
            = obj.pos, obj.h, obj.w, obj.l, obj.ry # mtr format
#             = obj.pos, obj.h, obj.w, obj.l, obj.ry # kitti
        total_gt_bbox_parameter_list.append(gt_boxes3d[k].tolist())

#     print(bboxes3d_[:,:3].shape)
    invalid_region_mask = train_dataset._get_invalid_region_mask(gt_boxes3d[:,:3])
    gt_boxes3d = gt_boxes3d[~invalid_region_mask,:]
    
    for k in range(len(gt_boxes3d)):
        gt_bbox_params = [gt_boxes3d[k, 5], gt_boxes3d[k, 3], gt_boxes3d[k, 4],
                        gt_boxes3d[k, 1], gt_boxes3d[k, 2], gt_boxes3d[k, 0],
                        gt_boxes3d[k, 6]]

        gt_bbox_params_list.append(gt_bbox_params)

    
    
    
    if gt_boxes3d.__len__() == 0:
        print('No gt object')
        continue

    # if(sample_id == 203):
    #     print(gt_boxes3d.shape)

    bboxes3d = mtr_utils.objs_to_boxes3d(obj_list)
#     invalid_region_mask = train_dataset._get_invalid_region_mask(bboxes3d_[:,:3])
#     bboxes3d = bboxes3d_[~invalid_region_mask,:]
    
    boxes_pts_mask_list = []

    bboxes3d_rotated_corners = mtr_utils.boxes3d_to_corners3d(bboxes3d)

    for i, bbox3d_corners in enumerate(bboxes3d_rotated_corners):
        box3d_roi_inds = mtr_utils.in_hull(point_cloud[:,:3], bbox3d_corners)
        boxes_pts_mask_list.append(box3d_roi_inds)
        total_number_of_points.append(np.sum(box3d_roi_inds))
    
    total_number_of_objects_per_scene.append(len(boxes_pts_mask_list))
    
    coor = point_cloud[:,[1,2,0]]
#     print(coor.shape)
#     print(np.array(bboxes3d_rotated_corners).shape)
    bboxes3d_rotated_corners_np  =np.array(bboxes3d_rotated_corners)
    npoints = bboxes3d_rotated_corners_np.shape[0] * bboxes3d_rotated_corners_np.shape[1]
#     print(np.reshape(bboxes3d_rotated_corners_np, (npoints,3)).shape)
    coor = np.concatenate([coor, np.reshape(bboxes3d_rotated_corners_np, (npoints,3))[:,[1,2,0]] ], axis=0)
#     print(coor.shape)
    
    
#     gt_bbox_params_list = []
    Converter.compile("raw_train_sample_bbox_corners_{}".format(sample_id), coors=coor, intensity=None,
                    bbox_params=gt_bbox_params_list)

#     # total_mask = boxes_pts_mask_list[0]
#     for k in range(boxes_pts_mask_list.__len__()):
#         pt_mask_flag = boxes_pts_mask_list[k]
#         # total_mask |= pt_mask_flag
#         # print("Sum of points: ", np.sum(pt_mask_flag))
#         if(np.sum(pt_mask_flag) == 0):
#             print("Zero Point Bounding Boxes")
#             continue
#         cur_pts = pts_rect[pt_mask_flag].astype(np.float32)
#         cur_pts_features = pts_features[pt_mask_flag].astype(np.float32)
#         sample_dict = {'sample_id': sample_id,
#                        'cls_type': obj_list[k].cls_type,
#                        'gt_box3d': gt_boxes3d[k],
#                        'points': cur_pts,
#                        'features': cur_pts_features,
#                        'obj': obj_list[k]}
#         # print(cur_pts.shape)
#         gt_database.append(sample_dict)
```
